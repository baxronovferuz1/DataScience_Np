```python
x=np.arange(0,4*np.pi, 0.1)
y=np.sin(x)
```

```python
plt.figure(figsize=(11,5))

plt.plot(x,y, linewidth=3, linestyle="-", color="b",marker="o")
plt.title("sin(x) function")
plt.xlabel("X line")
plt.ylabel("Y line")
plt.grid()
plt.show()
```

```python
x=np.arange(0,4*np.pi, 0.1)
sin=np.sin(x)
cos=np.cos(x)

plt.figure(figsize=(12,5))

#sin(x)
plt.plot(x,sin, linewidth=3, linestyle="-", color="g",marker="o")
plt.plot(x,cos, linewidth=3, linestyle="-", color="b",marker="*")

plt.title("sin(x) and cos(x) function graph")
plt.legend(["sin(x)","cos(x)"], loc=1, fontsize="x-large")
plt.grid()
plt.savefig("graph.png")
plt.show()


```
